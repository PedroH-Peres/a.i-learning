```python
import pandas as pd
import matplotlib.pylab as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python

dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('dados/AirPassengers.csv', parse_dates = ['Month'], index_col = 'Month', date_parser = dateparse)
ts = data['#Passengers']
```

```python
plt.plot(ts)
```

```python
decomposition = seasonal_decompose(ts)
```

```python
trending = decomposition.trend
trending
```

```python
seasonal = decomposition.seasonal
seasonal
```

```python
aleatory = decomposition.resid
aleatory
```

```python
plt.plot(seasonal)
```

```python
plt.plot(trending)
```

```python
plt.plot(aleatory)
```

```python
plt.subplot(4,1,1)
plt.plot(ts, label='Original')
plt.legend(loc = 'best')

plt.subplot(4,1,2)
plt.plot(trending, label='Trending')
plt.legend(loc = 'best')

plt.subplot(4,1,3)
plt.plot(seasonal, label='Seasonal')
plt.legend(loc = 'best')

plt.subplot(4,1,4)
plt.plot(aleatory, label='Aleatory')
plt.legend(loc = 'best')
```
